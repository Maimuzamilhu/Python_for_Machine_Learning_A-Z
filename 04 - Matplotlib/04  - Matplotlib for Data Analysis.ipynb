 # IMDB Movie Data Analysis

```python
movies = pd.read_csv("04 - IMDB-Movie-Data.csv" , nrows = 50)
movies.head()
```

```python
movies.columns
```

```python
movies.fillna(0 , inplace = True)
```

```python
plt.figure(figsize = (10,6))
plt.plot(movies["Rank"] , movies["Revenue (Millions)"])
plt.show()
```

```python
idx = movies["Revenue (Millions)"].idxmax()
movies.iloc[idx]
```

###### Line Styling and Putting Labels

```python
plt.figure(figsize = (10,3))

plt.plot(movies["Rank"] , movies["Rating"] , color = "r" , linestyle = "--",
            linewidth = 2 , marker = "o", markersize = 10 , markerfacecolor = "y" ,
             markeredgecolor = 'b' , markeredgewidth = 1)  #r for red

plt.xlabel("Rank")
plt.ylabel("Rating")
plt.title("Rank vs Rating")
```

```python
plt.figure(figsize = (10 , 3))
plt.scatter(movies["Rank"] , movies["Rating"])


plt.figure(figsize = (10 , 3))
plt.bar(movies["Rank"] , movies["Rating"] , color = "r" , width = 0.9)
plt.plot(movies["Rank"] , movies["Rating"] , color = "g" , linestyle = "--",
            linewidth = 2 , marker = "o", markersize = 10 , markerfacecolor = "y" ,
             markeredgecolor = 'b' , markeredgewidth = 1)

plt.figure(figsize = (3 , 10))
plt.barh(movies["Rank"] , movies["Rating"] , color = "b")

plt.figure(figsize = (10,3))
plt.hist(movies["Rating"] , bins = 20 , color="r" ,alpha = 0.5)

plt.figure(figsize = (3,10))
plt.hist(movies["Rating"] , bins = 20 , color="r" ,alpha = 0.5,
        orientation = "horizontal")

plt.figure(figsize=(10,3))
plt.boxplot(movies["Rating"] , patch_artist =True)

plt.show()
```

# Subplot

```python
plt.figure(figsize=(6,6))
plt.subplot(1,2,1)
plt.barh(movies["Rank"] , movies["Rating"] , color = "b")

plt.subplot(1,2,2)
plt.hist(movies["Rating"] , bins = 20 , color="r" ,alpha = 0.9,
        orientation = "horizontal")

plt.show()
plt.tight_layout()
```

 ###### 2x2 subplot

```python
plt.figure(figsize=(8,6))
#1-------------------------
plt.subplot(2,2,1)
plt.plot(movies["Rank"] , movies["Rating"] , color = "g" , linestyle = "--",
            linewidth = 2 , marker = "o", markersize = 10 , markerfacecolor = "y" ,
             markeredgecolor = 'b' , markeredgewidth = 1)

plt.xlabel("Rank")
plt.ylabel("Rating")
plt.title("Rank vs Rating")

#2------------------------------------
plt.subplot(2,2,2)
plt.plot(movies["Rank"] , movies["Revenue (Millions)"] , color = "b" , linestyle = "--",
            linewidth = 2 , marker = "o", markersize = 10 , markerfacecolor = "r" ,
             markeredgecolor = 'b' , markeredgewidth = 1)

plt.xlabel("Rank")
plt.ylabel("Revenue (Millions)")
plt.title("Rank vs Revenue (Millions)")

#3-------------------------------------------------
plt.subplot(2,2,3)
plt.plot(movies["Rank"] , movies["Metascore"] , color = "k" , linestyle = "--",
            linewidth = 2 , marker = "o", markersize = 10 , markerfacecolor = "y" ,
             markeredgecolor = 'b' , markeredgewidth = 1)

plt.xlabel("Rank")
plt.ylabel("Metascore")
plt.title("Rank vs Metascore")

#4-------------------------------------------------
plt.subplot(2,2,4)
plt.plot(movies["Rank"] , movies["Votes"] , color = "y" , linestyle = "--",
            linewidth = 2 , marker = "o", markersize = 10 , markerfacecolor = "b" ,
             markeredgecolor = 'y' , markeredgewidth = 1)

plt.xlabel("Rank")
plt.ylabel("Votes")
plt.title("Rank vs Votes")

plt.tight_layout()
```

##### Subplots

```python
#2x2 subplots

fig , ax = plt.subplots(nrows = 2 , ncols = 2 ,figsize = (8,6))

ax[0][0].plot(movies["Rank"] , movies["Rating"] , color = "g" , linestyle = "--",
            linewidth = 2 , marker = "o", markersize = 10 , markerfacecolor = "y" ,
             markeredgecolor = 'b' , markeredgewidth = 1)

ax[0][0].set_xlabel("Rank")
ax[0][0].set_ylabel("Rating")
ax[0][0].set_title("Rank vs Rating")


ax[1][1].plot(movies["Rank"] , movies["Votes"] , color = "g" , linestyle = "--",
            linewidth = 2 , marker = "o", markersize = 10 , markerfacecolor = "y" ,
             markeredgecolor = 'b' , markeredgewidth = 1)

ax[1][1].set_xlabel("Rank")
ax[1][1].set_ylabel("Votings")
ax[1][1].set_title("Rank vs Votings")

plt.tight_layout()
```

## Creating a Zoomed Sub-Figure of a Figure

```python
fig = plt.figure(figsize = (10,3))
ax1 = fig.add_axes([0,0 ,1,1]) #original
ax2 = fig.add_axes([0.1 , 0.15 ,0.2 ,0.3]) #zoomed


ax1.plot(movies["Rank"] , movies["Rating"] , color = "g" , linestyle = "--",
            linewidth = 2 , marker = "o", markersize = 10 , markerfacecolor = "y" ,
             markeredgecolor = 'b' , markeredgewidth = 1)


ax1.set_xlabel("Rank")
ax2.set_ylabel("Rating")
ax1.set_title("Rank vs Rating")

##plot zoomed Subplots

ax2.plot(movies["Rank"] , movies["Rating"] , color = "g" , linestyle = "--",
            linewidth = 2 , marker = "o", markersize = 10 , markerfacecolor = "y" ,
             markeredgecolor = 'b' , markeredgewidth = 1)


ax2.set_xlim(10,25) #set zoomed
ax2.set_ylim(4,9)


ax2.set_xlabel("Rank")
ax2.set_ylabel("Rating")
ax2.set_title("Rank vs Rating (zoomed)")

plt.show()
```

 ### xlim and ylim , legend , grid , xticks , yticks

```python
plt.figure(figsize = (10,3))
plt.plot(movies["Rank"] , movies["Rating"] , linestyle = "--" , color="r",
        label = "Rating")

plt.xlim(0,30)
plt.ylim(4,10)
plt.xlabel("Rank")
plt.ylabel("Rating")
plt.title("Rank vs Rating")
plt.legend()

plt.show()
```

```python
plt.figure(figsize = (10,3))

plt.plot(movies["Rank"] , movies["Rating"] , linestyle = "--" , color="r",
        label = "Rating")
plt.plot(movies["Rank"] , movies["Rating"]+1 , linestyle = "--" , color="g",
        label = "Rating +1")

plt.xlim(0,30)
plt.ylim(4,10)
plt.xlabel("Rank")
plt.ylabel("Rating")
plt.title("Rank vs Rating")
plt.legend()
plt.grid()

plt.xticks(np.arange(0,30,2))

plt.show()
```
